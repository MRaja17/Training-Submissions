We'll create a tiny orders table and compute totals and summaries.

```python

import pandas as pd

# Create a tiny dataset
orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104, 105],
    "customer": ["A", "B", "A", "C", "B"],
    "item": ["Pen", "Notebook", "Pencil", "Pen", "Pen"],
    "qty": [3, 2, 5, 1, 4],
    "unit_price": [1.50, 4.00, 0.80, 1.50, 1.50]
})

# Total price per row
orders = orders.assign(total=orders["qty"] * orders["unit_price"])

# Filter pens only
pens = orders.loc[orders["item"] == "Pen"]

# Group by customer to get total spent
by_customer = orders.groupby("customer", as_index=False)["total"].sum()

orders, pens, by_customer
```
