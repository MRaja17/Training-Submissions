# Intermediate 1 — Apply Pandas on a Real Dataset (Iris)

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the iris CSV generated alongside this notebook
df = pd.read_csv(r"/mnt/data/pandas_fundamentals_challenges/data_iris.csv")

# Quick inspection
print(df.head())
print("\nInfo:")
print(df.info())

# Summary statistics by species
group_means = df.groupby("species").mean(numeric_only=True)
print("\nMeans by species:\n", group_means)

# Class balance
balance = df["species"].value_counts().rename_axis("species").reset_index(name="count")
print("\nClass counts:\n", balance)

# Simple histogram example
df["sepal length (cm)"].hist()
plt.title("Distribution of Sepal Length (cm)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```


**Explanation (key results):**
- We inspected the first rows and schema to understand the columns and dtypes.
- `groupby("species").mean()` shows how flower measurements differ across species.
- `value_counts()` provides the class distribution.
- The histogram reveals the spread of sepal lengths and potential multi‑modal patterns by species (if plotted per species).
